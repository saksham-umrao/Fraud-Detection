```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
cwd = os.getcwd()
print(cwd)
```

```python
data_path = os.path.join(cwd, 'dataset')
print(data_path)
```

```python
print("loading the datasets...")

transaction_df = pd.read_csv(os.path.join(data_path, "train_transaction.csv"))
identity_df = pd.read_csv(os.path.join(data_path, "train_identity.csv"))

print("merging the datasets on TransactionID...")

df = transaction_df.merge(identity_df, on="TransactionID", how="left")

print("saving the merged dataset as 'train_merged.csv'...")

df.to_csv(os.path.join(data_path, "train_merged.csv"), index=False)

print("the merged dataset saved successfully.")
```
